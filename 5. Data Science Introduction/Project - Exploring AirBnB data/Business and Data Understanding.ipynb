## Looking into Seattle AirBnB Open Data


The dataset is available at https://www.kaggle.com/airbnb/seattle.

There are 3 files in the dataset.
- listings.csv:  This contain information about each listing.
- calendar.csv: Contains price of a listing on a particular date, if it is available.
- reviews.csv: Contains user reviews of listings.

Let's explore them.

## Looking into calendar.csv:

```python
calendar_df = pd.read_csv('./seattle/calendar.csv')
print(calendar_df.shape)
calendar_df.head()
```

```python
nan_percentage_per_column = calendar_df.isnull().mean()
columns_with_nan = nan_percentage_per_column[nan_percentage_per_column > 0]
columns_with_nan
```

```python
calendar_df[calendar_df['price'].notna()]['available'].value_counts()
```

```python
calendar_df[calendar_df['price'].isna()]['available'].value_counts()
```

We can see the rows which have 'price' as null are booked, and are marked 'f' in 'available' column.

```python
grouped_df = calendar_df.groupby('listing_id')['price']
grouped_df.groups
```

```python
grouped_df.get_group(9419).value_counts()
```

```python
 # 6606
listings_df.loc[listings_df['id'] == 9419, 'price']
```

We can see from here that each listing can have different prices on different dates. The prices mentioned in listings.csv are maybe the latest price as per the last scrape.
There are NaN values only in price column, indicating the listing is booked.
